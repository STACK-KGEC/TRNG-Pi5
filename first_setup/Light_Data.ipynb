```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
light_data_path = "/home/admin/Python/light_data.csv"

data = pd.read_csv("/home/admin/Python/light_data.csv")
data['timestamp'] = pd.to_datetime(data['timestamp'])
data
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
dpi = 400
plt.figure(figsize=(12, 8), dpi=dpi)
plt.plot(data['timestamp'], data['lux'], label='Light Intensity (lux)', color='darkviolet')
plt.xlabel("Time Stamp", fontsize=12, weight="bold")
plt.ylabel("Light Intensity (in Lux)", fontsize=12, weight="bold")
plt.title("Light Intensity (in Lux)", fontsize=16, weight="bold")
plt.legend()
plt.grid(linestyle="--", alpha=0.8)
plt.show()
```
